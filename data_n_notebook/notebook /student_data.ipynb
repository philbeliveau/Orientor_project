```python
import random
from faker import Faker
from datetime import datetime, timedelta
fake = Faker()

# Generating synthetic data
num_rows = len(df_students)

# We'll generate random picks for each row:
df_students["Favourite movie"] = [random.choice(fake_movies) for _ in range(num_rows)]
df_students["Favourite Book"] = [random.choice(fake_books) for _ in range(num_rows)]
df_students["Role Model"] = [random.choice(fake_roles) for _ in range(num_rows)]
df_students["Learning Style"] = [random.choice(learning_styles) for _ in range(num_rows)]
df_students['Interests']= [random.choice(fake_interests) for _ in range(num_rows)] 

start = 9
new_user = num_rows + start

# Generate user ID 
df_students['ID'] = [id for id in range(start, new_user)]
df_students['user_id'] = [random.randint(start, new_user - 1) for _ in range(num_rows)]


df_students.head()
```

```python
# -------------------------------------------------------------
# 3) Create a Synthetic "Users" DataFrame
# -------------------------------------------------------------
# We'll generate the same number of new user rows.
# We'll skip IDs 1..8, so we start from 9..(9+num_rows-1)
list_of_ids = list(range(start, start + num_rows))

emails = []
hashed_passwords = []
created_ats = []

for _ in list_of_ids:
    # Generate email with Faker
    # Or you could parse from "Name" if you want them to match
    email = fake.email()
    
    # Fake hashed password, or you could store hashed versions if you have a hashing function
    # We'll just do a simple random hex for demonstration
    fake_hash = fake.sha256(raw_output=False)
    
    # Random creation date within last 365 days
    random_days_ago = random.randint(0, 365)
    created_date = datetime.now() - timedelta(days=random_days_ago)
    
    emails.append(email)
    hashed_passwords.append(fake_hash)
    created_ats.append(created_date.isoformat())

df_users = pd.DataFrame({
    "ID": list_of_ids,
    "email": emails,
    "hashed_password": hashed_passwords,
    "created_at": created_ats
})
```
